#### Вам нужно помочь секретарю автоматизировать работу. Для этого нужно написать программу, которая будет на основе хранимых данных исполнять пользовательские команды.

Исходные данные имеют следующую структуру:


    перечень всех документов
    documents = [
     {'type': 'passport', 'number': '2207 876234', 'name': 'Василий Гупкин'},
     {'type': 'invoice', 'number': '11-2', 'name': 'Геннадий Покемонов'},
     {'type': 'insurance', 'number': '10006', 'name': 'Аристарх Павлов'}]

перечень полок, на которых хранятся документы (если документ есть в documents, то он обязательно должен быть и в directories)

    directories = {
     '1': ['2207 876234', '11-2'],
     '2': ['10006'],
     '3': []}
     
Общие требования к программе:


* код должен быть грамотно декомпозирован (каждая функция отвечает за свою конкретную задачу, дублирующийся функционал переиспользуется, а его код не повторяется);
* в коде отсутствуют глобальные переменные (за исключением documents и directories);
* пользовательский ввод обрабатывается в цикле while до тех пор, пока пользователь явно не завершит программу (вводом команды "q").

### Задание 1

#### Пункт 1. Пользователь по команде "p" может узнать владельца документа по его номеру

#### Пункт 4. Пользователь по команде "ads" может добавить новую полку

Примеры работы:

2. Введите команду: ads

#### Пункт 5. Пользователь по команде "ds" может удалить существующую полку из данных (только если она пустая)¶

Примеры работы:

1. Введите команду: ds
    
   Введите номер полки: 3

Результат:
Полка удалена. Текущий перечень полок: 1, 2.

2. Введите команду: ds
   
   Введите номер полки: 1

Результат:
На полке есть документа, удалите их перед удалением полки. Текущий перечень полок: 1, 2, 3.

3. Введите команду: ds
   
   Введите номер полки: 4

Результат:
Такой полки не существует. Текущий перечень полок: 1, 2, 3.



```python
# Основная функция, в которой по значению параметра происходит определенное действие. 
def main (parameter):
    if parameter == 'p':
        number = input('Введите номер документа: ') 
        command_p(number)
        
    elif parameter == 'q':
        print('Работа алгоритма остановлена')
        
    elif parameter == 's':
        number = input('Введите номер документа: ')
        if command_s(number) != None: 
            print('Документ хранится на полке:', command_s(number))
        else: 
            print('Результат: Документ не найден в базе')
            
    elif parameter == 'l':
        command_l()
        
    elif parameter == 'ads':
        number = input('Введите номер полки:')
        command_ads(number)
    
    elif parameter == 'ds':
        number = input('Введите номер полки:')
        command_ds(number)
    
    else:
        print('Данная команда не предусмотрена')
```

```python
# Функция поиска владельца документа по номеру этого документа
def command_p(number1):
    k=0
    for line in documents:
        if number1 == line['number']:
            print('Владелец документа:', line['name'])
            k = k+1
    if k == 0:
        print('Результат: Документ не найден в базе')
```

```python
# Функция поиска полки по номеру документа
def command_s(number1):
    for line, name in directories.items():
        if number1 in name:
            return line
```

```python
# Функция вывода полной информации по всем документам
def command_l():
    for line in documents:
        print('№:',line['number'], ' тип:', line['type'], ' владелец:', line['name'], ' полка хранения:', command_s(line['number']))
```

```python
# Функция добавления новой полки
def command_ads(number1):
    if number1 in directories.keys():
        print('Такая полка уже существует.Текущий перечень полок:') 
        output()
    else:
        directories[number1] = ''  
        print('Полка добавлена. Текущий перечень полок:')
        output()
        print(directories[number1])
```

```python
# Функция удаления пустой полки
def command_ds(number1):
    if number1 not in directories.keys():
        print('Такой полки не существует.Текущий перечень полок:') 
        output()
    elif directories[number1] == [] or directories[number1] == '':
        del(directories[number1])
        print('Полка удалена. Текущий перечень полок:')
        output()
    else:
        print('На полке есть документы, удалите их перед удалением полки. Текущий перечень полок:')
        output()
```

```python
# Вывод перечня полок
def output():
    for key in directories.keys():
        print(key)
```

```python
# Основной код программы
documents = [
 {'type': 'passport', 'number': '2207 876234', 'name': 'Василий Гупкин'},
 {'type': 'invoice', 'number': '11-2', 'name': 'Геннадий Покемонов'},
 {'type': 'insurance', 'number': '10006', 'name': 'Аристарх Павлов'}]

directories = {
 '1': ['2207 876234', '11-2'],
 '2': ['10006'],
 '3': []}

command_q = ''
while command_q != 'q':
    command_q = input('Введите команду: ')
    main(command_q)
    print('')
```
