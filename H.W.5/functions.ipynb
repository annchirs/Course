```python
# Основная функция, в которой по значению параметра происходит определенное действие. 
def main (parameter):
    if parameter == 'p':
        command_p()
       
    elif parameter == 'q':
        print('Работа алгоритма остановлена')
        
    elif parameter == 's':
        number_d = input('Введите номер документа: ')
        if command_s(number_d) != None: 
            print('Документ хранится на полке:', command_s(number_d))
        else: 
            print('Результат: Документ не найден в базе')
            
    elif parameter == 'l':
        command_l()
        
    elif parameter == 'ads':
        command_ads()
    
    elif parameter == 'ds':
        command_ds()
        
    elif parameter == 'ad':
        command_ad()
        
    elif parameter == 'ad': 
        command_ad()   
        
    elif parameter == 'd':
        command_d()
    
    elif parameter == 'm':
        command_m()
        
    else:
        print('Данная команда не предусмотрена')
```

```python
# Функция поиска владельца документа по номеру этого документа
def command_p():
    number = input('Введите номер документа: ') 
    k=0
    for line in documents:
        if number == line['number']:
            print('Владелец документа:', line['name'])
            k = k+1
    if k == 0:
        print('Результат: Документ не найден в базе')
```

```python
# Функция поиска полки по номеру документа
def command_s(number1):
    for line, name in directories.items():
        if number1 in name:
            return line
```

```python
# Функция вывода полной информации по всем документам
def command_l():
    for line in documents:
        print('№:',line['number'], ' тип:', line['type'], ' владелец:', line['name'], ' полка хранения:', command_s(line['number']))
```

```python
# Функция добавления новой полки
def command_ads():
    number = input('Введите номер полки:')
    if number in directories.keys():
        print('Такая полка уже существует.Текущий перечень полок:') 
        output()
    else:
        directories[number] = []  
        print('Полка добавлена. Текущий перечень полок:')
        output()
```

```python
# Функция удаления пустой полки
def command_ds():
    number = input('Введите номер полки:')
    if number not in directories.keys():
        print('Такой полки не существует.Текущий перечень полок:') 
        output()
    elif directories[number] == []:
        del(directories[number])
        print('Полка удалена. Текущий перечень полок:')
        output()
    else:
        print('На полке есть документы, удалите их перед удалением полки. Текущий перечень полок:')
        output()
```

```python
# Вывод перечня полок
def output():
    for key in directories.keys():
        print(key)
```

```python
# Добавление нового документа
def command_ad():
    number_d = input('Введите номер документа:')
    type_d = input('Введите тип документа:')
    owner = input('Введите владельца документа:')
    number_p = input('Введите полку для хранения:')
    print('')
    if number_p not in directories.keys():
        print('Такой полки не существует, добавьте полку командой ads. Текущий список полок:')
        output()
    elif command_s(number_d):
        print('Документ с таким номером уже хранится на полке. Текущий список документов:')
        command_l()
    else:
        documents.append({'type':type_d, 'number':number_d, 'name':owner})
        directories[number_p].append(number_d)
        print('Документ добавлен. Текущий список документов:')
        command_l()
```

```python
# Функция удаления документа из данных
def command_d():
    number = input('Введите номер документа:')
    if command_s(number) == None:
        print('Документа с таким номером не существует. Текущий список документов:')
        command_l()
    else:
        i = 0
        while i != len(documents):
            if documents[i]['number'] == number:
                del(documents[i])
                print('Документ удален. Текущий список документов: ')
                command_l()
            else:
                i += 1
        for elem in directories.values():
            k = 0
            while k != len(elem):
                if elem[k] == number:
                    del(elem[k])
                else:
                    k += 1
```

```python
# Функция для перемещения документа с одной полки на другую.
def command_m():
    number_d = input('Введите номер документа:')
    number_p = input('Введите номер полки:')
    if number_p not in directories.keys():
        print('Такой полки не существует, добавьте полку командой ads. Текущий список полок:')
        output()
    elif command_s(number_d) == None:
        print('Документа с таким номером не существует. Текущий список документов:')
        command_l()
    else:
        for elem in directories.values():
            k = 0
            while k != len(elem):
                if elem[k] == number_d:
                    del(elem[k])
                else:
                    k += 1
        directories[number_p].append(number_d)
        print('Документ перемещен. Текущий список документов:')
        command_l()
```

```python
# Основной код программы
documents = [
 {'type': 'passport', 'number': '2207 876234', 'name': 'Василий Гупкин'},
 {'type': 'invoice', 'number': '11-2', 'name': 'Геннадий Покемонов'},
 {'type': 'insurance', 'number': '10006', 'name': 'Аристарх Павлов'}]

directories = {
 '1': ['2207 876234', '11-2'],
 '2': ['10006'],
 '3': []}

command_q = ''
while command_q != 'q':
    command_q = input('Введите команду: ')
    main(command_q)
    print('')
    
```
