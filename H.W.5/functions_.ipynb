```python
# Основная функция, в которой по значению параметра происходит определенное действие. 
def main (document, directory):
    commands_list = {
        'l': command_l,
        'ad': command_ad,
        'd': command_d,
        'm': command_m
    }
    parameter = ''
    while parameter != 'q':
        parameter = input('Введите команду: ')
        
        if parameter == 'p':
            command_p(document)
            
        elif parameter == 's':
            number_d = input('Введите номер документа: ')
            if command_s(number_d, directory) != None: 
                print('Документ хранится на полке:', command_s(number_d, directory))
            else: 
                print('Результат: Документ не найден в базе')
        
        elif parameter == 'ads':
            command_ads(directory)
        
        elif parameter == 'ds':
            command_ds(directory)
        
        elif parameter in commands_list:
            commands_list[parameter](document, directory)
            
        elif parameter and parameter != 'q':
            print('Данная команда не предусмотрена')
        print('')    
    print('Работа алгоритма остановлена')
```

```python
# Функция поиска владельца документа по номеру этого документа
def command_p(doc):
    number = input('Введите номер документа: ') 
    k=0
    for line in doc:
        if number == line['number']:
            print('Владелец документа:', line['name'])
            k = k+1
    if k == 0:
        print('Результат: Документ не найден в базе')
```

```python
# Функция поиска полки по номеру документа
def command_s(number1,direct):
    for line, name in direct.items():
        if number1 in name:
            return line
```

```python
# Функция вывода полной информации по всем документам
def command_l(doc,direct):
    for line in doc:
        print('№:',line['number'], ' тип:', line['type'], ' владелец:', line['name'], ' полка хранения:', command_s(line['number'],direct))
```

```python
# Функция добавления новой полки
def command_ads(direct):
    number = input('Введите номер полки:')
    if number in direct.keys():
        print('Такая полка уже существует.Текущий перечень полок:') 
        output(direct)
    else:
        direct[number] = []  
        print('Полка добавлена. Текущий перечень полок:')
        output(direct)
```

```python
# Функция удаления пустой полки
def command_ds(direct):
    number = input('Введите номер полки:')
    if number not in direct.keys():
        print('Такой полки не существует.Текущий перечень полок:') 
        output(direct)
    elif direct[number] == []:
        del(direct[number])
        print('Полка удалена. Текущий перечень полок:')
        output(direct)
    else:
        print('На полке есть документы, удалите их перед удалением полки. Текущий перечень полок:')
        output(direct)
```

```python
# Вывод перечня полок
def output(direc):
    for key in direc.keys():
        print(key)
```

```python
# Добавление нового документа
def command_ad(doc, direct):
    number_d = input('Введите номер документа:')
    type_d = input('Введите тип документа:')
    owner = input('Введите владельца документа:')
    number_p = input('Введите полку для хранения:')
    print('')
    if number_p not in direct.keys():
        print('Такой полки не существует, добавьте полку командой ads. Текущий список полок:')
        output(direct)
    elif command_s(number_d, direct):
        print('Документ с таким номером уже хранится на полке. Текущий список документов:')
        command_l(doc, direct)
    else:
        doc.append({'type':type_d, 'number':number_d, 'name':owner})
        direct[number_p].append(number_d)
        print('Документ добавлен. Текущий список документов:')
        command_l(doc, direct)
```

```python
# Функция удаления документа из данных
def command_d(doc, direct):
    number = input('Введите номер документа:')
    if command_s(number, direct) == None:
        print('Документа с таким номером не существует. Текущий список документов:')
        command_l(doc, direct)
    else:
        i = 0
        while i != len(doc):
            if doc[i]['number'] == number:
                del(doc[i])
                print('Документ удален. Текущий список документов: ')
                command_l(doc, direct)
            else:
                i += 1
        remove(number, direct)
```

```python
# Удаление документа с полки
def remove(number1, direc):
    for elem in direc.values():
            k = 0
            while k != len(elem):
                if elem[k] == number1:
                    del(elem[k])
                else:
                    k += 1
```

```python
# Функция для перемещения документа с одной полки на другую.
def command_m(doc, direct):
    number_d = input('Введите номер документа:')
    number_p = input('Введите номер полки:')
    if number_p not in direct.keys():
        print('Такой полки не существует, добавьте полку командой ads. Текущий список полок:')
        output(direct)
    elif command_s(number_d,direct) == None:
        print('Документа с таким номером не существует. Текущий список документов:')
        command_l(doc, direct)
    else:
        remove(number_d, direct)
        direct[number_p].append(number_d)
        print('Документ перемещен. Текущий список документов:')
        command_l(doc, direct)
```

```python
# Основная программа
documents = [
 {'type': 'passport', 'number': '2207 876234', 'name': 'Василий Гупкин'},
 {'type': 'invoice', 'number': '11-2', 'name': 'Геннадий Покемонов'},
 {'type': 'insurance', 'number': '10006', 'name': 'Аристарх Павлов'}]

directories = {
 '1': ['2207 876234', '11-2'],
 '2': ['10006'],
 '3': []}

main(documents, directories)
```
