### Задание 1

Даны 2 переменных, в которых хранятся строки произвольной длины: phrase_1 и phrase_2.
Напишите код, который проверяет какая из этих строк длиннее.

Примеры работы программы:

- phrase_1 = 'Насколько проще было бы писать программы, если бы не заказчики'
- phrase_2 = '640Кб должно хватить для любых задач. Билл Гейтс (по легенде)'

Результат:
Фраза 1 длиннее фразы 2

```python
phrase_1 = 'Насколько проще было бы писать программы, если бы не заказчики'
phrase_2 = '640Кб должно хватить для любых задач. Билл Гейтс (по легенде)'

if len(phrase_1)>len(phrase_2):
    print('Результат: Фраза 1 длиннее фразы 2')
elif len(phrase_2)>len(phrase_1):
    print('Результат: Фраза 2 длиннее фразы 1')
else:
    print('Результат: Фразы одинаковые по длине')
```

Пример работы программы:

year = 2020

Результат:
Високосный год

```python
year = 2020

if (year % 4==0) and (year % 100 != 0) or (year % 400 == 0):
    print('Результат: Високосный год')
else:
    print ('Результат: Обычный год')
```

### Задание 5 (необязательное)

Дана переменная, в которой хранится шестизначное число (номер проездного билета). Напишите программу, которая будет определять, является ли данный билет "счастливым". Билет считается счастливым, если сумма первых трех цифр совпадает с суммой последних трех цифр номера.

Примеры работы программы:

number = 123456

Результат:
Несчастливый билет

```python
number = 123213

sum1 = (number // 100000) + (number // 10000 % 10) + (number // 1000 % 10) #Сумма первых 3-х цифр
sum2 = (number % 10) + (number % 100 // 10) + (number % 1000 // 100) #Сумма последних 3-х цифр
if sum1 == sum2:
    print('Результат: Счастливый билет')
else:
    print('Результат: Несчастливый билет')
```

### Задание 6 (необязательное)
Напишите программу, которая сможет вычислять площади трех фигур (круг, треугольник и прямоугольник). Тип фигуры запрашиваем через пользовательский ввод, после чего делаем запрос характеристик фигуры:

* если пользователь выбрал круг, запрашиваем его радиус,
* если треугольник – длины трех его сторон;
* если прямоугольник – длины двух его сторон.

Пример работы программы:

Введите тип фигуры:
Круг

Введите радиус круга:
10

Результат:
Площадь круга: 314.16

```python
pi = 3.14159265
figure = input('Введите тип фигры:')
if figure == 'Круг':
    r = int(input('Введите радиус круга:'))
    s_circle = pi * (r**2)
    print('Площадь круга:',s_circle)
elif figure == 'Треугольник':
    a = int(input('Введите длину стороны а:'))
    b = int(input('Введите длину стороны b:'))
    c = int(input('Введите длину стороны c:'))
    p = (a + b + c)/2
    s_triangle = (p * (p-a) * (p-b) * (p-c))**(1/2)
    print('Площадь треугольника:',s_triangle)
elif figure == 'Прямоугольник':
    a = int(input('Введите длину стороны а:'))
    b = int(input('Введите длину стороны b:'))
    s_rectangle = a * b
    print('Площадь прямоугольника:',s_rectangle)
else:
    print('Неправильный тип фигуры')
```
