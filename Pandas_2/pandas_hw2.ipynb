### Задание 1
Напишите функцию, которая классифицирует фильмы из материалов занятия по следующим правилам:
- оценка 2 и меньше - низкий рейтинг
- оценка 4 и меньше - средний рейтинг
- оценка 4.5 и 5 - высокий рейтинг

Результат классификации запишите в столбец class

```python
import pandas as pd
data = pd.read_csv('ratings.csv')
data.head()
```

```python
def raiting_class(row):
    if row.rating >= 4.5:
        return 'высокий рейтинг'
    elif 2 < row.rating < 4.5:
        return 'средний рейтинг'
    elif row.rating <= 2:
        return 'низкий рейтинг'
```

```python
data['class'] = data.apply(raiting_class, axis = 1)
data.head()
```

### Задание 2
Используем файл keywords.csv.

Необходимо написать гео-классификатор, который каждой строке сможет выставить географическую принадлежность определенному региону. Т. е. если поисковый запрос содержит название города региона, то в столбце ‘region’ пишется название этого региона. Если поисковый запрос не содержит названия города, то ставим ‘undefined’.

Правила распределения по регионам Центр, Северо-Запад и Дальний Восток:

geo_data = {'Центр': ['москва', 'тула', 'ярославль'], 'Северо-Запад': ['петербург', 'псков', 'мурманск'], 'Дальний Восток': ['владивосток', 'сахалин', 'хабаровск']}

Результат классификации запишите в отдельный столбец region.

```python
data1 = pd.read_csv('keywords.csv')
data1.tail()
```

```python
def geo_data(row):
    if 'москва' in row['keyword'] or 'тула' in row['keyword'] or 'ярославль' in row['keyword']:
        return 'Центр'
    elif 'петербург' in row['keyword'] or 'псков' in row['keyword'] or 'псков' in row['keyword']:
        return 'Северо-Запад'
    elif 'владивосток' in row['keyword'] or 'сахалин' in row['keyword'] or 'хабаровск' in row['keyword']:
        return 'Дальний Восток'
    else:
        return 'undefined'
```

```python
data1['region'] = data1.apply(geo_data, axis=1)
data1['region'].value_counts()
```

### Задание 3 (бонусное)

Есть мнение, что “раньше снимали настоящее кино, не то что сейчас”. Ваша задача проверить это утверждение, используя файлы с рейтингами фильмов из прошлого домашнего занятия (файл movies.csv и ratings.csv из базы https://grouplens.org/datasets/movielens/). 
Т. е. проверить верно ли, что с ростом года выпуска фильма его средний рейтинг становится ниже.


    

```python
movies = pd.read_csv('movies.csv')
ratings = pd.read_csv('ratings.csv')

ratings_movies = movies.merge(ratings)
ratings_movies.head()
```

```python
"""В переменную years запишите список из всех годов с 1950 по 2010."""

years = [str(i+1950) for i in range(61)]
```

```python
"""2) Напишите функцию production_year, которая каждой строке из названия фильма выставляет год выпуска. 
Не все названия фильмов содержат год выпуска в одинаковом формате, поэтому используйте следующий алгоритм:
    1 для каждой строки пройдите по всем годам списка years
    2 если номер года присутствует в названии фильма, то функция возвращает этот год как год выпуска
    3 если ни один из номеров года списка years не встретился в названии фильма, то возвращается 1900 год"""

def production_year(row):
    for year in years: 
        if year in row.title:
            return  year
    return '1900'
```

```python
"""3) Запишите год выпуска фильма по алгоритму пункта 2 в новый столбец ‘year’"""
ratings_movies['year'] = ratings_movies.apply(production_year, axis=1)
ratings_movies.head()
```

```python
"""4) Посчитайте средний рейтинг всех фильмов для каждого значения столбца ‘year’ и отсортируйте результат по убыванию рейтинга"""
ratings_movies = ratings_movies.groupby('year').mean().reset_index()[['year', 'rating']]
ratings_movies = ratings_movies.sort_values(by='rating', ascending = False)
ratings_movies.head()
```
